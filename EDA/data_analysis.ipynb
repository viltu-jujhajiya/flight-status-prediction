```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from pyspark.sql import SparkSession
from glob import glob
```

```python
spark = (
    SparkSession.builder \
    .master("local[*]") \
    .appName("Flight Status Prediction") \
    .config("spark.sql.shuffle.partitions", "30") \
    .config("spark.ui.showConsoleProgress", "false") \
    .getOrCreate()
)

spark.sparkContext.setLogLevel("WARN")
```

```python
datapath = r"/Users/viltujujhajiya/Desktop/Career/Flight-Status-Prediction/data"
csv_paths = []
for path in glob(os.path.join(datapath, "*.csv")):
    if "flights" in path.lower():
        csv_paths.append(path)

csv_paths
```

```python
total_files = 0
for path in csv_paths:
    csv = spark.read.csv(path, inferSchema = True, header = True)
    files = csv.count()
    total_files += files
    print(f"Total Number of flight detailes in {path.split('/')[-1]}: {files}")
```
